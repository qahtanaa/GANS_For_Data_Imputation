# Testing on German dataset

```python
#First, save your raw data set that needs to be imputed
german_data = pd.read_csv("/content/TGAIN/rawData/german/german_data_preprocessed.csv", sep = ',')
german_data = german_data.drop(columns=['Unnamed: 0'])
german_data.head(5)

```

```python
#Split into a train- and test set
complete_rows_german = german_data.dropna() #432 complete rows
complete_rows_german = shuffle(complete_rows_german, random_state=0)
test_set_complete = complete_rows_german[:50] #50 test rows
train_set = complete_rows_german[50:] #382 training rows


#Create a mask and randomly remove 30% of the test set 
data_m = np.random.random(test_set_complete.shape) < 0.3 #30% missing data
test_set_to_impute = test_set_complete.mask(data_m) #test set with roughly 30% randomly missing values

total_df_german = pd.concat([train_set, test_set_to_impute], ignore_index=True, sort=False) 
```

```python
#Start imputing using the provided framework
completed_df = data_imputation_completer(total_df_german, nr_epochs=300, distinct_ratio_threshold=0.01, generator_lr=0.0002, discriminator_lr=0.0002, verbose = False)
```

```python
#Before imputation
test_set_to_impute
```

```python
#After Imputation
test_set_imputed = completed_df.tail(50)
test_set_imputed
```

```python
#Evaluation of imputations
evaluate_imputations(train_set, test_set_complete, data_m, test_set_imputed, 0.01) #Prediction error
```
